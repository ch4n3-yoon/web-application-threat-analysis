```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
countries = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = countries.plot(column="continent", legend=True, categorical=True)
ax.set_title("세계 지도")
ax.set_axis_off()
```

```python
df = pd.read_csv("./result1.csv")

grouped = df.groupby(by=["country_name"], as_index=False)
grouped_df = grouped["country_code"].count()
grouped_df["count"] = list(grouped_df["country_code"])
grouped_df.head()
```

```python
import plotly.express as px

fig = px.choropleth(grouped_df, locations="country_name",
                    locationmode="country names",
                    color="count", 
                    hover_name="country_name",
                    color_continuous_scale=px.colors.sequential.Reds)
fig.show()
```
